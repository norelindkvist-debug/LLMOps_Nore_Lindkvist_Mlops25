```python
with open("mlops_raw.txt", "r") as file:
    mlops_text = file.read()

mlops_text[:150]
```

```python
from pydantic import BaseModel, Field

class Education(BaseModel):
    educaition_name: str = Field(description="Make sure to find this information in the text")
    program_description: str = Field(description="A 3-5 sentence summary about the program")
    career_options: list[str]
    courses: list[str]
    length: float = Field(description="Total study length of the program")
```

```python
from pydantic_ai import Agent

education_extractor_agent = Agent(
    "openrouter:arcee-ai/trinity-mini:free",
    system_prompt="""
    You are an exper in Swedish education, especially on Yrkeshögskola (YH). 
    Based on a text about an education, you will extract out relevant information.

    Make sure to use same language as the text. 
"""
)

result = await education_extractor_agent.run(mlops_text, output_type=Education)
result.output
```
